You have a matrix $X$ of size (10,5) filled with random values. Print the following :
* Minimum value and index where it occured, for each column
* Maximum value and the index where it occured, for each column
* Minimum value and index where it occured, for each row
* Maximum value and index where it occured, for each row
* Minimum value and index where it occured, for the entire matrix
* Maximum value and index where it occured, for the entire matrix

````
np.random.seed(0)
X = np.random.rand(10,5)

print("Minimal value is {} and index is {}".format(np.amin(X,axis=0),np.argmin(X,axis=0)))
print("Maximum value is {} and index is {}".format(np.amax(X,axis=0),np.argmax(X,axis=0)))

print("Minimal value is {} and index is {}".format(np.amin(X,axis=1),np.argmin(X,axis=1)))
print("Maximum value is {} and index is {}".format(np.amax(X,axis=1),np.argmax(X,axis=1)))

xDash = X.flatten()
print("minimum value is {} and index is {}".format(min(xDash),np.argmin(xDash)))
print("maximum value is {} and index is {}".format(max(xDash),np.argmax(xDash)))
````

Find and print the following for the matrix X: 
* Mean of each row and column
* Standard deviation of each row and column
* Mean of the entire matrix
* Standard deviation of entire matrix

````
print("Mean of each row {} and column is {}".format(np.mean(X,axis=1),np.mean(X,axis=0)))
print("Standard deviation of each row {} and column is {}".format(np.std(X,axis=1),np.std(X,axis=0)))

print("mean of the entire matrix is {}".format(X.mean()))
print("Standard deviation of the entire matrix is {}".format(X.std()))
````

````
# Initial imports
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
````

Download dataset from: http://biostat.mc.vanderbilt.edu/wiki/pub/Main/DataSets/titanic3.csv
Read the data and display the first 5 rows

````
#Download dataset from: http://biostat.mc.vanderbilt.edu/wiki/pub/Main/DataSets/titanic3.csv

file_url = "http://biostat.mc.vanderbilt.edu/wiki/pub/Main/DataSets/titanic3.csv"

data = pd.read_csv(file_url)
data.head(5)
````

### Preprocessing of the Dataset



---


**1. Convert the categorical values of columns 'sex', 'embarked', 'pclass' into dummy values and add them into titanic_data_new DataFrame**




---



````
titanics_data_new = pd.get_dummies(data,columns=['sex','embarked','pclass'])
titanics_data_new
````



---


**2. Fill the missing data in Age and Fare columns with the mean value of the respected column then add these columns to titanic_data_new**


---



````
titanics_data_new['age'].isna()
````

````
titanics_data_new['fare'].isna()
````

````
print(titanics_data_new['age'].mean())
print(titanics_data_new['fare'].mean())
````

````
titanics_data_new['age'] = titanics_data_new['age'].fillna((titanics_data_new['age'].mean()))
titanics_data_new['fare'] = titanics_data_new['fare'].replace(0,titanics_data_new['fare'].mean())
titanics_data_new.head(10)
````

###Normalization


---



1. Normalize the columns which have numerical data



---



````
# normalize with min max normalization

titanics_data_new["fare"]=((titanics_data_new["fare"]-titanics_data_new["fare"].min())/(titanics_data_new["fare"].max()-titanics_data_new["fare"].min()))
titanics_data_new["age"]=((titanics_data_new["age"]-titanics_data_new["age"].min())/(titanics_data_new["age"].max()-titanics_data_new["age"].min()))

titanics_data_new
````



---
2. Drop the original age, fare columns
3. Append the normalized columns to the titanic_data_new DataFrame.



---



````
# Drop the original age and fare column
data.drop(['age','fare'],axis=1)
data
````

````
# Append the normalized columns to the titanic_data_new DataFrame.
data['age'] = titanics_data_new["age"]
data['fare'] = titanics_data_new["fare"]
data
````

Plot the sine function with domain in $[-3\pi,3\pi]$. Label the axes appropriately. Add title $\sin(x)$. Use latex formatting while adding the title. Set the plot size to (10,10)

````

fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 10
fig_size[1] = 10
plt.rcParams["figure.figsize"] = fig_size

x = np.arange(-3*np.pi,3*np.pi,0.1)   # start,stop,step
y = np.sin(x)
plt.title('Sine wave')
plt.xlabel('time')
plt.ylabel('y = sin(x)')
plt.axhline(y=0, color='k')
plt.plot(x,y)
plt.show()
````

Plot $\sin(x)$ and its spectrum, $\cos(x)$ and its spectrum in a single plot, in the domain $[-2\pi,2\pi]$. Label the plot appropriately.

````
x = np.arange(-2*np.pi,2*np.pi,0.1)  
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y,x,z)
plt.xlabel('x values from -2pi to 2pi')  
plt.ylabel('sin(x) and cos(x)')
plt.title('Plot of sin and cos from -2pi to 2pi')
plt.legend(['sin(x)', 'cos(x)'])      
plt.show()
````
